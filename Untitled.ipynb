```python
import pandas as pd 
df=pd.read_csv("D:186/PY/CAR DETAILS FROM CAR DEKHO.csv")

df.info()
```

```python
df.head(10)
```

```python
df['Company']=df['name'].apply(lambda x: x.split(' ')[0])
```

```python
df['year_old']=df['year'].apply(lambda x: 2024-x)
```

```python
df.drop(columns=['name','year_old'],inplace=True)
```

```python
df
```

```python
df=pd.get_dummies(df,drop_first=True)
```

```python
df
```

```python
x=df.drop(columns=['selling_price'])
y=df['selling_price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=20)
```

```python
from sklearn.linear_model import LinearRegression
m=LinearRegression()
m.fit(x_train,y_train)
y_pred=m.predict(x_test)
```

```python
pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
```

```python
from sklearn.metrics import r2_score,mean_squared_error
print(r2_score(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

```python
import pandas as pd 

data=pd.read_csv("D:/186/PY/Ice_cream selling data.csv")

data.head()
```

```python
sns.regplot(data=data,x=data.columns[0],y='Ice Cream Sales (units)')
```

```python
x=data[['Temperature (C)']]
y=data['Ice Cream Sales (units)']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=20)
```

```python
from sklearn.linear_model import LinearRegression
m=LinearRegression()
m.fit(x_train,y_train)
y_pred=m.predict(x_test)
```

```python
pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
```

```python
from sklearn.metrics import r2_score,mean_squared_error
print(r2_score(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf=PolynomialFeatures(degree=4)
x_poly=pf.fit_transform(x)

```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_poly,y,test_size=0.2,random_state=20)
```

```python
from sklearn.linear_model import LinearRegression
m=LinearRegression()
m.fit(x_train,y_train)
y_pred=m.predict(x_test)
```

```python
pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
```

```python
from sklearn.metrics import r2_score,mean_squared_error
print(r2_score(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```
